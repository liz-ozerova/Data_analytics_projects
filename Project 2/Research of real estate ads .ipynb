### 1.2.5 Анализ вариантов заполнения столбца "населенный пункт"

Для более точного анализа возможного расположения - была создана таблица с квартирами с неуказанным местом расположения. 

```python
# печать столбцов с отсутствующими данными по расположению парка, водоема и аэропорта
# no_data_list - список с параметрами таблицы, связанные с географическим положением 
no_data_list = ['locality_name', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest']

# функция для составление строки условий выборки
def string_assembly(string):
    conditions = ''
    len_string = len(string) - 1
    for i in range(len_string):
        conditions += string[i] + '.isnull()' + ' and '
    conditions += string[len_string] + '.isnull()'
    return conditions

# составление условий
# информация о количестве объявлений без какой-либо информации о локации
conditions_no_data = string_assembly(no_data_list)
advt_no_location = advt.query(conditions_no_data)
print('Количество записей без какой-либо информации о локации', len(advt_no_location))
display(advt_no_location)
print('\n', 'Записи без города')
table = advt.query('locality_name.isnull()')
print('Количество записей без указания города', len(table))
```

Строк с полностью отсутствующей информации о расположении квартиры в том числе косвенной -всего 8, без указания города - 49. Можно попытаться предположить, что самые дороие из этих квартир расположены в Санкт-Петербурге,  однако могут быть квартиры в пригородах Санкт-Петербурга в элитных новостройках

```python
print ('\033[1m' + 'Таблица с данными без указания города' + '\033[0m')
display(table.head(15))
table['meter_price'] = table['last_price'] / table['total_area']
data_StP = advt[advt['locality_name'] == 'Санкт-Петербург']
data_StP['meter_price_StP'] = data_StP['last_price'] / data_StP['total_area']
mean_price_StP = data_StP['meter_price_StP'].mean()
print('Средняя цена за метр в Санкт-Петербурге, {: .2f}'.format(mean_price_StP), '\n')
print ('\033[1m' + 'Таблица сводная с данными без указания локации, для которых средняя за м2 цена больше средней для Санкт-Петербурга ' + '\033[0m')
part_table = table.loc[:, ['cityCenters_nearest', 'meter_price', 'last_price', 'total_area', 'rooms']]
display(part_table[part_table['meter_price'] >= mean_price_StP].sort_values(by='cityCenters_nearest', ascending=False))
```
